## Загрузка библиотек и исходных данных

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
HOUSING_PATH = "data/housing.csv"

housing = pd.read_csv(HOUSING_PATH)
housing.head()
```

## 1. Масштабирование 

а. Нормализовать признак - median_income

```python
housing.median_income_norm = (housing.median_income - housing.median_income.mean())/(housing.median_income.max()-housing.median_income.min())
housing.head()
```

б. Привести признак - population к масштабу от 0 до 100

```python
scaler = MinMaxScaler()
housing[['population']] = scaler.fit_transform(housing[['population']]) * 100
housing.head()
```

## 2. Заполнение пропусков:

```python
housing['total_bedrooms'] = housing[['ocean_proximity', 'population']].groupby('ocean_proximity').transform(lambda x: x.fillna(x.mean()))
housing.info()
```
